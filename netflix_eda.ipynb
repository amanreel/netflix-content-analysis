```python
import pandas as pd

# Load the CSV from the Colab path
df = pd.read_csv('/content/netflix_titles.csv')
df.head()

# Convert 'date_added' to datetime safely
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year and month safely
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

# Fill missing values in key columns
df['country'] = df['country'].fillna('Unknown')
df['director'] = df['director'].fillna('No Director Info')

# Preview to confirm it's working
df[['title', 'type', 'date_added', 'year_added', 'country']].head()

import seaborn as sns
import matplotlib.pyplot as plt

# Plot distribution of content type
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type', palette='Set2')
plt.title("Content Type Distribution")
plt.xlabel("Type")
plt.ylabel("Count")
plt.tight_layout()

# Save the plot to Colab
plt.savefig("type_distribution.png")
plt.show()

# Bar plot of content added per year
plt.figure(figsize=(10,5))
df['year_added'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title("Titles Added Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.tight_layout()
plt.savefig("yearly_trend.png")
plt.show()

# Get top 10 countries
top_countries = df['country'].value_counts().head(10)

# Plot
top_countries.plot(kind='bar', color='coral')
plt.title("Top 10 Content-Producing Countries")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.tight_layout()
plt.savefig("top_countries.png")
plt.show()

from wordcloud import WordCloud

# Join all director names into one string
text = ' '.join(df['director'])

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Plot
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Frequent Directors on Netflix")
plt.tight_layout()
plt.savefig("wordcloud_directors.png")
plt.show()

from google.colab import files

# Download each plot one by one
files.download("type_distribution.png")

files.download("yearly_trend.png")
python
Copy
Edit
files.download("top_countries.png")
python
Copy
Edit
files.download("wordcloud_directors.png")

files.download("top_countries.png")


files.download("wordcloud_directors.png")
```
